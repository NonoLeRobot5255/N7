Au début, on observe que l'augmentation de la longueur des chemins considérés entraîne l'ajout de liens entre les villes. Cependant, le nombre de liens finit par se stabiliser. 
Cela s'explique par le fait que nous considérons seulement l'existence d'un chemin d'une longueur maximale de k. 
Or dans un graphe connexe de n sommets, augmenter k ajoutera des arêtes entre les sommets, mais lorsque k atteint ou dépasse n, tous les sommets seront déjà connectés. Par conséquent, augmenter davantage k n'ajoute plus d'arêtes.
Dans notre cas, les villes reliés entre elles forment plusieurs sous graphes connexes.

<p style="text-align: justify; font-size: 16px; color: #4682b4;">
<b>Q1)</b> Quelles sont les conditions sur les paires successives des sommets d’une chaîne pour que celle-ci appartienne au graphe ?
</p>

</div>

Soit $G=(V,E)$ un graphe connexe et $C = (v_1, v_2, v_3, \ldots, v_k)$ une chaîne.
Pour que la chaîne $C$ appartienne au graphe, il que que chaque paire successive de sommets $(v_i, v_{i+1})$ dans la chaîne, soit une arrête du graphe. 
Donc il faut que pour $1\leq i < k$, $(v_i, v_{i+1}) \in E$.
Une autre condition est l'existance des sommets considérés.
Ainsi, la vérification de l'appartenance d'une chaîne à un graphe repose sur la vérification de l'existence des sommets dans $V$ et des arêtes correspondantes dans $E$.

<div style="background-color: #f0f8ff; padding: 20px; border-radius: 10px; border: 1px solid #87cefa; font-family: 'Arial', sans-serif; line-height: 1.6;">

<p style="text-align: justify; font-size: 16px; color: #4682b4;">
<b>Q2)</b> Écrire la fonction <code style="color: #ff4500;">possedechaine(A, chaîne)</code> qui permet de déterminer si une chaîne donnée appartient à un graphe <i>G</i> de matrice d'adjacence <i>A</i>.
</p>

</div>

```matlab
%A faire

%Les tests sont valides uniquement pour d=500
tester_possedechaine(A);

function chain_exists = possedechaine(adjMatrix, c)
    % POSSEDECHAINE Vérifie si une chaîne est présente dans le graphe
    % G est la matrice d'adjacence
    % c est le vecteur représentant la chaîne
    % La fonction retourne true si la chaîne est présente, sinon false

    % Nombre de nœuds dans la chaîne
    n = length(c);
    
    % Initialisation de la variable de retour
    chain_exists = true;
    
    % Boucle pour vérifier chaque paire de nœuds dans la chaîne
    for i = 1:(n-1)
        % Si il n'y a pas d'arête entre deux nœuds consécutifs, la chaîne n'existe pas
        if adjMatrix(c(i), c(i+1)) == 0
            chain_exists = false;
            break;
        end
    end
end

%Test
function tester_possedechaine(A)
    % TESTER_POSSEDECHAINE Fonction pour valider la fonction possedechaine
    % Cette fonction charge les chaînes et les valeurs de vérité, exécute possedechaine pour
    % chaque chaîne, et utilise assert pour valider si les résultats sont conformes aux attentes.

    % Chemin vers le dossier contenant les chaînes
    chainsDir = 'donnees/tests/possedechaine';
    files = dir(fullfile(chainsDir, 'chain_*.mat'));
    
    % Charger les valeurs de vérité
    truthValues = load('donnees/tests/possedechaine/reponses/truthValues.mat');
    truths = truthValues.truthValues;  % Assumer que les valeurs sont dans la variable 'truthValues'
    % Boucle pour tester chaque chaîne
    for k = 1:length(files)
        % Charger la chaîne courante
        chainFile = fullfile(chainsDir, files(k).name);
        data = load(chainFile);
        chain = data.chain; % Assumer que les chaînes sont stockées dans la variable 'chain'
        
        % Extraire le numéro de la chaîne depuis le nom du fichier
        [~, name, ~] = fileparts(files(k).name);
        chainNumber = sscanf(name, 'chain_%d');
        
        % Obtenir le résultat attendu de la liste de vérités
        expectedResult = truths(chainNumber);
        
        % Calculer le résultat en utilisant possedechaine
        result = possedechaine(A, chain);
        % Utiliser assert pour vérifier le résultat
        if expectedResult
            resultStr = 'true';
        else
            resultStr = 'false';
        end

        assert(result == expectedResult, ...
            ['Échec du test pour la chaîne : ', files(k).name, '. Attendu ', ...
            resultStr, ', obtenu ', mat2str(result), '.']);
    end
    
    disp('Tous les tests ont été passés avec succès.');
end
```

<p style="text-align: justify; font-size: 16px; color: #4682b4;">
On s'intéresse maintenant en particulier aux chaînes eulériennes. Une chaîne eulérienne est une chaîne empruntant une fois et une fois seulement chaque arête du graphe.
</p>

</div>

<div style="background-color: #c8e6c9; padding: 20px; border-radius: 10px; border: 1px solid #388e3c; font-family: 'Arial', sans-serif; line-height: 1.6;">

<p style="text-align: justify; font-size: 16px; color: #2e7d32;">
    <br><b>Rappel </b>
    <details>
        <summary></summary>
        <b>Théorème d'Euler</b> : Un graphe admet une chaîne eulérienne si et seulement si, en exceptant les sommets isolés, il est connexe et si ses sommets sont tous de degré pair sauf au plus deux.
    </details>
</p>

</div>

<div style="background-color: #f0f8ff; padding: 20px; border-radius: 10px; border: 1px solid #87cefa; font-family: 'Arial', sans-serif; line-height: 1.6;">

<p style="text-align: justify; font-size: 16px; color: #4682b4;">
<b>Q1)</b> Implémenter à l'aide des fonctions matricielles une méthode <code style="color: #ff4500;">isEulerien</code> pour déterminer si un graphe possède une chaîne eulérienne.
</p>

</div>

```matlab
%A faire
test_isEulerian();

function is_eulerian = isEulerian(adjMatrix)
    % IS_EULERIAN vérifie si un graphe déjà connecté est eulérien
    % Selon le théorème d'Euler, un graphe est eulérien si tous les sommets ont un degré pair.
    
    % On vérifie la connexité pour les sommets reliés par des arrêtes
    if ~verifierConnexite(adjMatrix)
        is_eulerian = false; % Affecter la valeur false à is_eulerian
        return;              % Quitter la fonction
    end

    % Compter le nombre de sommets de degré impair
    nb_impairs = sum(mod(sum(adjMatrix, 2), 2) == 1);

    % Vérifier si au plus 2 sommets ont un degré impair
    is_eulerian = nb_impairs < 3;
end

function estConnexe = verifierConnexite(G)
    % Cette fonction vérifie si un graphe est connexe
    % Entrée:
    %   G : une matrice d'adjacence du graphe
    % Sortie:
    %   estConnexe : un booléen (true ou false)
    
    % Nombre de sommets dans le graphe
    n = size(G, 1);
    degres = sum(G, 2);

    % Filtrer les sommets avec au moins une arête
    sommets_actifs = find(degres > 0);

    % Initialisation de la liste des sommets visités
    visite = false(1, n);
    
    % Démarrage de la visite par le premier sommet actif, s'il existe
    if isempty(sommets_actifs)
        estConnexe = true; % Un graphe sans arêtes est trivialement connexe
        return;
    end
    queue = sommets_actifs(1);
    visite(queue) = true;

    while ~isempty(queue)
        current = queue(1);
        queue(1) = [];
        
        % Trouver tous les sommets non visités adjacents à 'current'
        for i = sommets_actifs'
            if G(current, i) == 1 && ~visite(i)
                visite(i) = true;
                queue(end+1) = i;  % Ajout du sommet à la queue
            end
        end
    end
    
    % Le graphe est connexe si tous les sommets actifs ont été visités
    estConnexe = all(visite(sommets_actifs));
end

% Test
function test_isEulerian()
    % TEST_IS_EULERIAN vérifie si les graphes représentés par des matrices d'adjacence sont eulériens
    % Cette fonction parcourt un dossier de matrices, charge chaque matrice, et teste si elle représente
    % un graphe eulérien en utilisant la fonction 'isEulerian'. Elle utilise les valeurs de vérité pré-enregistrées
    % pour valider les résultats de 'isEulerian'.

    % Chemin vers le dossier contenant les matrices
    matrixDir = 'donnees/tests/isEulerian';
    files = dir(fullfile(matrixDir, '*.mat'));
    
    % Supposons que les valeurs de vérité sont stockées dans un fichier MAT séparé
    truthValues = load('donnees/tests/isEulerian/reponses/truthValues.mat');
    truths = truthValues.truthValues;  % Supposons que la variable soit nommée 'truthValues'
    
    % Parcourir chaque fichier et tester la fonction 'isEulerian'
    for k = 1:length(files)
        % Charger la matrice
        matrixFile = fullfile(matrixDir, files(k).name);
        data = load(matrixFile);
        matrix = data.adj_matrix;  % Supposons que la matrice soit enregistrée sous le nom de variable 'adj_matrix'
        
        % Extraire le numéro de la matrice à partir du nom de fichier
        [~, name, ~] = fileparts(files(k).name);
        matrixNumber = sscanf(name, 'adj_matrix_%d');
        
        % Obtenir le résultat attendu à partir de la liste de vérités
        expectedResult = truths(matrixNumber);
        
        % Calculer le résultat en utilisant 'isEulerian'
        result = isEulerian(matrix);
        
        % Vérifier le résultat
        if expectedResult
            resultStr = 'true';
        else
            resultStr = 'false';
        end
        assert(result == expectedResult, ...
            ['Échec du test pour la matrice : ', files(k).name, '. Attendu ', ...
            resultStr, ', obtenu ', mat2str(result), '.']);
    end
    
    % Afficher un message si tous les tests sont passés avec succès
    disp('Tous les tests ont été passés avec succès.');
end
```

Dans le cas actuel, le graphe n'est pas connexe, il n'est donc pas possible de passer par toutes les arrêtes à partir d'une chaîne.
On ne peut donc pas créer de chaîne eulérienne et donc le graphe n'est pas eulérien.

</div>

<p style="text-align: justify; font-size: 16px; color: #4682b4;">
On cherche à identifier les valeurs de <i>d</i> pour lequel le graphe admet une chaîne eulérienne.
</p>
</div>

<div style="background-color: #f0f8ff; padding: 20px; border-radius: 10px; border: 1px solid #87cefa; font-family: 'Arial', sans-serif; line-height: 1.6;">

<p style="text-align: justify; font-size: 16px; color: #4682b4;">
<b>Q1)</b>Proposer une méthode qui détermine pour toutes les valeurs de portée <i>d</i> (dans <i>R</i>) si <i>G(d)</i> possède une chaîne eulérienne.
</p>
</div>

On peut procéder ainsi pour avoir toutes les valeurs de portée <i>d</i> (dans $\mathbb{R}$) si <i>G(d)</i> possède une chaîne eulérienne : 

1. **Extraction des distances distinctes**
On commence par parcourir toutes les arêtes du graphe $D$ pour extraire les distances distinctes $d_1, d_2, \ldots, d_{m}$ en les triant par ordre croissant. Cela nous permet de ne pas avoir à itérer sur tout $\mathbb{R}$ en de ne considérer que les moments où le nombre d'arrêtes change. En effet $\forall d_a,d_b,\in[d_k,d_{k+1}[$, $G(d_a)=G(d_b)$ donc la propriété le caractère eulérien de $G(d_a)$ et $G(d_b)$ ne dépend que du fait $G(d_k)$ soit eulérien ou non.

2. **Construction incrémentale des graphes $G(d)$** :
On va donc commencer par initaliser $d$ à 0. On va ensuite itérer sur $d_1, d_2, \ldots, d_{m}$. A chaque étape, on construit $G(d_k)$ et on regarde si il est eulérien. On prolonge ensuite ce résultat sur l'intervalle $[d_k,d_{k+1}[$. Dans le cas particuler de $G(d_m)$, on prolonge le résultat sur $[d_m,+\infty[$

Pour la valeur de portée minimale $d=175$ rendant le graphe eulérien, cette même valeur correspond également à la première situation où deux villes sont reliées. Étant donné qu'il existe une unique paire de villes pour chaque distance, dans ce cas, deux villes sont connectées par une arête, tandis que toutes les autres restent des sommets isolés. Ainsi, le graphe est eulérien.

Lorsque $d=3907$, tous les sommets sont connectés les uns aux autres, à l'exception de deux, notés $n$ et $n-1$, qui ne sont pas directement connectés entre eux, bien qu'ils soient tous deux reliés à tous les autres sommets. Avec 27 villes, chaque sommet, à l'exception de $n$ et $n-1$, a un degré de 26, tandis que $n$ et $n-1$ ont un degré de 25.
